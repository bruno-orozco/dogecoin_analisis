```python
# Para renombrar las etiquetas de nuestro DataFrame utilizamos la funci√≥n rename
dogecoin.rename(columns={'Date': 'Fecha',
                         'Open': 'Precio de apertura',
                         'High': 'Precio m√°s alto',
                         'Low': 'Precio m√°s bajo',
                         'Close': 'Precio de cierre'}, inplace=True)

dogecoin
```

```python
dogecoin.columns
```

```python
#La funci√≥n drop nos ayudar√° a eliminar elementos que deseemos en nuestro DataFrame
dogecoin.drop(['Adj Close'], axis=1, inplace=True)

dogecoin
```

```python
#Podemos agregar m√°s columnas para nuestro an√°lsis
dogecoin['Diferencia'] = dogecoin['Precio de cierre'] - dogecoin['Precio de apertura']
dogecoin['Rango'] = dogecoin['Precio m√°s alto'] - dogecoin['Precio m√°s bajo']
dogecoin['Promedio'] = ((dogecoin['Precio de cierre'] + dogecoin['Precio de apertura'] + dogecoin['Precio m√°s alto'] + dogecoin['Precio m√°s bajo']) / 4)
dogecoin
```

```python
#Para agregar nuestra primera gr√°fica utilizamos Boxplot, para m√°s detalles consultar en:
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.boxplot.html

distribucion = dogecoin[['Precio de apertura', 'Precio de cierre', 'Precio m√°s alto', 'Precio m√°s bajo', 'Diferencia', 'Rango', 'Promedio']]
grafica_distribucion = distribucion.boxplot(color = 'gray', figsize = [14, 6])
grafica_distribucion.set_title('Dogecoin datos hist√≥ricos gr√°fica de disperci√≥n')
```

```python
#Mi an√°lisis es este: ......
#Con la funci√≥n plot, podemos agregar una gr√°fica lineal
#Tambi√©n con ella podremos agregar m√°s graficos, para m√°s informaci√≥n: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html
dogecoin = dogecoin.set_index(['Fecha'])
volatilidad = dogecoin[['Precio m√°s alto', 'Precio m√°s bajo']]
grafica_volatilidad = volatilidad.plot(kind = 'line', color = ('gold', 'black'), figsize = [14, 4], fontsize = 10)
```

```python
#Con esta funci√≥n quitar√© el √≠ndice de fecha y lo regresar√© a las col
dogecoin = dogecoin.reset_index('Fecha')
dogecoin
```

```python
#Trnasformo la columna Fecha en tipo de dato datetime64

dogecoin['Fecha'] = pd.to_datetime(dogecoin['Fecha'])
dogecoin.dtypes
```

```python
#Con esta funci√≥n agrupar√© mis datos por semana, el argumento freq equivale en este caso a 1 week, es decir una semana
#https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases

doge2021 = dogecoin.copy(deep = True)
#doge2021 = doge2021.groupby(pd.Grouper(key = 'Fecha', freq = '1w')).sum()
doge2021 = doge2021[dogecoin['Fecha'].isin(pd.date_range('2021-01-01','2021-07-11'))]
doge2021 = doge2021.set_index(['Fecha'])
doge2021
```

```python
#Una vez que he agrupado mis datos por semana aplicar√© una vez m√°s mi c√≥digo pero ahora absorver√° el nuevo agrupamiento que he hecho por semanas
volatilidad = doge2021[['Precio m√°s alto', 'Precio m√°s bajo']]
grafica_volatilidad = volatilidad.plot(kind = 'line', color = ('gold', 'black'), figsize = [14, 4], fontsize = 10)
grafica_volatilidad.set_title('Volatilidad del Dogecoin en el 2021')
```

```python
distribucion = doge2021[['Precio de apertura', 'Precio de cierre', 'Precio m√°s alto', 'Precio m√°s bajo', 'Diferencia', 'Rango', 'Promedio']]
grafica_distribucion = distribucion.boxplot(color = 'gray', figsize = [14, 6])
grafica_distribucion.set_title('Dogecoin 2021')
```

```python
#Agrego una funci√≥n y creo un programa ramificado, si se cumple la primera condici√≥n
#el resultado en la nueva columna An√°lisis ser√° El precio es bajo, sino se cumple, el 
#resultado ser√° El precio es muy alto
def reglas(row):
  if row['Promedio'] < .005:
    return 'El precio es bajo'
  else:
    return 'El precio es muy alto'

dogecoin['An√°lisis'] = dogecoin.apply(reglas, axis=1)
dogecoin
```

```python
#https://plotly.com/python/linear-fits/
doge2021 = doge2021.reset_index(['Fecha'])
regresion_volume = px.scatter(doge2021, x='Fecha', y='Volume', trendline="ols")
regresion_volume.show()
```
